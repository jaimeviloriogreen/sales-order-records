```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importación del archivo Excel

```python
#Carga de archivo Excel
df_ventas = pd.read_excel("sales-records.xlsx")

df_ventas[ df_ventas["Zona"]== "África" ]
```

## 1. Determinar la cantidad y el porcentaje de las ventas realizadas según la zona

```python
zonas =  list( set( df_ventas["Zona"] ) )

frecuencia_ventas_zonas = [df_ventas[ df_ventas["Zona"] == zonas[n] ]["Zona"].count() for n in range(len(zonas))]
ventas_zonas = pd.DataFrame({"Zonas":zonas, "Ventas realizadas":frecuencia_ventas_zonas})
```

```python
total = pd.Series(frecuencia_ventas_zonas).sum()

ventas_zonas = ventas_zonas.sort_values("Ventas realizadas", ascending=False)

ventas_zonas["Porcentaje"] = (ventas_zonas["Ventas realizadas"] / total) * 100
ventas_zonas["Acumulativo"] = ventas_zonas["Porcentaje"].cumsum()

## Agregar el símbolo de porcentaje 
ventas_zonas["Porcentaje"] = [f"{val:.2f}%" for val in ventas_zonas["Porcentaje"]]
ventas_zonas["Acumulativo"] = [f"{val:.2f}%" for val in ventas_zonas["Acumulativo"]]
```

```python
ventas_zonas.index = range(1, len(ventas_zonas) + 1)

## Estilizado para alinearlo hacia el centro
ventas_zonas_styled = ventas_zonas.style.set_properties(**{'text-align': 'center'})
ventas_zonas_styled.set_table_styles([dict(selector='th', props=[('text-align', 'center')])])

ventas_zonas_styled
```

```python
# Gráfica de barra de las ventas realizadas por zona

plt.figure(figsize=(7, 5))

x = ventas_zonas["Zonas"]
y = ventas_zonas["Ventas realizadas"]

plt.bar(x,y, color=["#2874A6", "#A565BD"])

plt.title("Ventas por zonas")
plt.xlabel("Zonas")
plt.ylabel("Ventas realizadas")
plt.xticks(rotation=35, ha='right')

plt.grid(axis = 'y', color = '#767676', linestyle = 'dashed')
plt.grid(axis = 'x', color = '#767676', linestyle = 'dashed')

for n in range(0, len(x)):
    value = y[n + 1]
    plt.text(n, value + 1, value, ha="center", va="bottom", fontsize=9, color="#333", fontweight="bold")

plt.show()
```

### Análisis
El **37.60%** de todas las ventas registradas se realizaron en **África**, siendo este continente el que mayor registro de ventas posee. Otro **24.00%** de las ventas se llevaron a cabo en **Europa**, y un **16.20%** de las ventas totales fueron realizadas en **Asia**; siendo estos continentes los que acapararon aproximadamente el **80% de todas las ventas registradas**, el otro 20%  de las ventas restantes fueron registradas en América, Australia y Oceanía.
